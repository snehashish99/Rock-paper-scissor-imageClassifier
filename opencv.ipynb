```python
import cv2 as cv
import numpy as np
import os
import matplotlib.pyplot as plt 
```

```python
datadir="/home/unknown/PycharmProjects/OpenCV/rps"
```

```python
categories=["rock","paper","scissors"]
```

```python
IMG_SIZE_X = 50
IMG_SIZE_Y = 50
```

```python
for cat in categories:
    path = os.path.join(datadir, cat)
    for img in os.listdir(path):
        img_arr = cv.imread(os.path.join(path, img), cv.IMREAD_GRAYSCALE)
        plt.imshow(img_arr, cmap="gray")
        plt.show()
        break
    break
img_arr = cv.resize(img_arr, (IMG_SIZE_X, IMG_SIZE_Y))
plt.imshow(img_arr,cmap="gray")
plt.show()
```

```python
training_data = []
def create_training_data():
    for cat in categories:
        path = os.path.join(datadir, cat)
        label = categories.index(cat)
        for img in os.listdir(path):
            try:
                img_arr = cv.imread(os.path.join(path, img), cv.IMREAD_GRAYSCALE)
                img_arr = cv.resize(img_arr, (IMG_SIZE_X, IMG_SIZE_Y))
                training_data.append([img_arr, label])
            except Exception as e:
                #print(e)
                pass
    return training_data
create_training_data()
```

```python
len(training_data)
```

```python
import random 
random.shuffle(training_data)
training_data
```

```python
X = []
y = []
for feature, label in training_data:
    X.append(feature)
    y.append(label)
X 
```

```python
X = np.array(X).reshape(-1, IMG_SIZE_X, IMG_SIZE_Y,1)
y = np.array(y)
```

```python
X
```

```python
y
```

```python
import pickle 
pickle_out = open("x.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```

```python
from tensorflow.keras.utils import normalize
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Conv2D, MaxPooling2D, Dense, Activation
```

```python
X = normalize(X, axis=1)
```

```python
model = Sequential()

model.add(Conv2D(64, (3,3), input_shape=X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(512, activation="relu"))

model.add(Dense(3, activation="softmax"))

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

model.fit(X, y, batch_size=64, epochs=20, validation_split=0.1)
```
